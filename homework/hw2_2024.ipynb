В этом домашнем задании вы будете работать с корпусом "ISOT Fake News". Корпус состоит из новостных статей, которым присвоена метка класса, указывающая, является ли статья фальшивой или нет.

Корпус доступен по ссылке: https://drive.google.com/file/d/1edF_YVEtk2Reuc4hg9mksnaIVyTzbLFC/view?usp=sharing .

Используйте часть train для обучения, часть val для валидации, а часть test — для тестирования.

Каждый бонусный пункт оценивается в 1 балл.
