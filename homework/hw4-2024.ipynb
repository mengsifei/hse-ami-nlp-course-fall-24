## ПРАВИЛА
1. Домашнее задание выполняется в группе до 4-х человек.
2. Домашнее задание оформляется в виде отчета либо в .pdf файле, либо в ipython-тетрадке. Также обязательно прикладывать ipynb-файл. 
3. Отчет должен содержать: нумерацию заданий и пунктов, которые вы выполнили, код решения, и понятное пошаговое описание того, что вы сделали. Отчет должен быть написан в академическом стиле, без излишнего использования сленга и с соблюдением норм русского языка.
4. Не стоит копировать фрагменты лекций, статей и Википедии в ваш отчет.
5. Отчеты, состоящие исключительно из кода, не будут проверены и будут автоматически оценены нулевой оценкой.
6. Плагиат и любое недобросовестное цитирование приводит к обнулению оценки. 

В данном домашнем задании будем использовать этот [датасет](https://www.kaggle.com/datasets/kaushal2896/english-to-german).

## Часть 1. [1 балл] Эксплоративный анализ

1. Проанализируйте распределение длины предложений для английских и немецких текстов. Визуализируйте результаты.
2. Обнаруживаются ли различия в использовании пунктуации? Например, в английской версии используется `.`, а в немецкой версии — `!`. Какие еще интересные наблюдения можно сделать в этом наборе данных?



## Часть 2. [5 баллов] Реализация модели Encoder-Decoder

1. Разделите корпус на обучающую и тестовую выборки в соотношении 90% к 10%.
2. Реализуйте модель Encoder-Decoder с механизмом внимания.
3. Обучите модель и оцените её качество с использованием метрики BLEU.
4. Визуализируйте веса внимания для двух пар предложений.
5. Обсудите, как механизм внимания улучшает интерпретируемость и производительность модели.


## Часть 3. [3 балла] Генерация с использованием T5

1. Импортируйте предобученную модель T5. Оцените её качество на выбранной вами метрике и объясните выбор данной метрики.
2. Проведите дообучение модели на данном корпусе. Сравните качество дообученной модели с моделью из предыдущего пункта.
