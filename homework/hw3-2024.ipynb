В этом домашнем задании вы будете работать с корпусами - [BiodivNER](https://bdj.pensoft.net/article/89481/instance/7788198/) (в разделе `Data Resources`) и [COPIOUS](https://pmc.ncbi.nlm.nih.gov/articles/PMC6351503/)(в разделе `Supplementary material 2`).

В корпусе BiodivNER используются 6 типов именованных сущностей: 
* ORGANISM (e.g. plants, fungi, bacteria)
* ENVIRONMENT (environments species live in)
* QUALITY (characteristics to be measured)
* MATERIAL (e.g. chemical compounds)
* PROCESS (re-occurring biological and physical processes)
* LOCATION (geographic location)
* DATA TYPE (research results, e.g. lidar data), допускаются вложенные сущности. 

В корпусе COPIOUS используются 5 типов именованных сущностей: 
* Geographical Location
* Person
* Temporal Expression
* Taxon
* Habitat, допускаются вложенные сущности.
